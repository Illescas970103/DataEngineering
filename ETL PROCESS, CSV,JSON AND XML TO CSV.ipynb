```python
#lets create our EXTRACT FUNCTION for csv
def Extract_from_csv(file_to_process):
    dataframe=pd.read_csv(file_to_process)
    return dataframe
```

```python
#lets create out extract function for jason files
def Extract_from_JSON(file_to_process):
    dataframe=pd.read_json(file_to_process,lines=True)
    return dataframe
```

```python
#Lets create our extract function for XML files
def Extract_from_XML(file_to_process):
    dataframe=pd.DataFrame(columns=['name','height','weight'])
    tree=ET.parse(file_to_process)
    root=tree.getroot()
    for person in root:
        name=person.find('name').text
        height=float(person.find('height').text)
        weight=float(person.find('weight').text)
        dataframe=dataframe.append({'name':name,'height':height,'weight':weight},ignore_index=True)
    return dataframe
        
```

```python
#COMPLETE EXTRACTION
def extract():
    extracted_data=pd.DataFrame(columns=['name','height','weight'])
    
    #extraction for csv files
    for csvfile in glob.glob("*.csv"):
        extracted_data=extracted_data.append(Extract_from_csv(csvfile),ignore_index=True)
        
    #Extraction for Json files
    for jsonfiles in glob.glob('*.json'):
        extracted_data=extracted_data.append(Extract_from_JSON(jsonfiles),ignore_index=True)
        
    #extraction for XML files
    for xmlfiles in glob.glob('*.xml'):
        extracted_data=extracted_data.append(Extract_from_XML(xmlfiles),ignore_index=True)
    
    return extracted_data
```
